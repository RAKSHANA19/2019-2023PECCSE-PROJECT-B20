After LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
var_mod = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
le = LabelEncoder()
for i in var_mod:
    df[i] = le.fit_transform(df[i]).astype(int)
```
